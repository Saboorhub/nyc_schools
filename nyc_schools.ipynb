# Read in the data

#### First, we are going to read the csv files.

```python
import pandas as pd
import numpy as np
import re 

# store the csv files in the data_files list to read them all into a dataframe
data_files = [
    "ap_2010.csv",
    "class_size.csv",
    "demographics.csv",
    "graduation.csv",
    "hs_directory.csv",
    "sat_results.csv"]

data = {}

for file in data_files: 
    df = pd.read_csv("schools/{}".format(file))
    data[file.replace(".csv", "")] = df
```

#### Second, let us read the text files now. 

```python
# there are two text files. 
all_survey = pd.read_csv("schools/survey_all.txt", delimiter="\t", encoding='windows-1252')
d75_survey = pd.read_csv("schools/survey_d75.txt", delimiter="\t", encoding='windows-1252')

# let us concatenate the above two dataframes into one
survey = pd.concat([all_survey, d75_survey], axis=0, sort=True)

survey['DBN'] = survey['dbn']

# we are going to remove the irrelevant fields. 
survey_fields = ["DBN", 
    "rr_s", 
    "rr_t", 
    "rr_p", 
    "N_s", 
    "N_t", 
    "N_p", 
    "saf_p_11", 
    "com_p_11", 
    "eng_p_11", 
    "aca_p_11", 
    "saf_t_11", 
    "com_t_11", 
    "eng_t_11", 
    "aca_t_11", 
    "saf_s_11", 
    "com_s_11", 
    "eng_s_11", 
    "aca_s_11", 
    "saf_tot_11", 
    "com_tot_11", 
    "eng_tot_11", 
    "aca_tot_11",
]

survey = survey.loc[:, survey_fields]
data['survey'] = survey
```

# Add DBN columns

```python
# so we are going to create a DBN column from the already existing dbn column
data['hs_directory']['DBN'] = data['hs_directory']['dbn']

# since we have a 0 before single digits in the DBN column, we need to pad our csd column
# of the class_size, so we are going to create a padded_csd column out of csd column

data['class_size']['padded_csd'] = data['class_size']['CSD'].apply(lambda x: str(x).zfill(2))
data['class_size']['DBN'] = data['class_size']['padded_csd']+data['class_size']['SCHOOL CODE']
```

```python
# let us take a look at the first three rows of the class size
data['class_size'].head(3)

# we see there are two new columns, padded_csd and DBN. 
```

# Convert columns to numeric

```python
# let us convert the following list of cols in sat_results from object to numeric
SAT_cols = ['SAT Math Avg. Score', 'SAT Critical Reading Avg. Score', 'SAT Writing Avg. Score']

for col in SAT_cols: 
    data['sat_results'][col] = pd.to_numeric(data['sat_results'][col], errors='coerce')
    
## note, we are using using the errors="coerce" attribute as there are some errors
```

```python
# let us create another column and we will call it sat_score
# it will hold the combined score from the SAT_cols

data['sat_results']['sat_score'] = data['sat_results'].loc[:, SAT_cols].sum(axis=1)
```

```python
## let us extract the latitude and longitude for each school

pat = r'\(.+\)'

def lon_lat(string, lon=True): 
    coord_list = re.findall(pat, string)
    coord_string = coord_list[0]
    return coord_string.split(",")[1].replace(")", "") if lon==True else coord_string.split(",")[0].replace("(", "")


data['hs_directory']['lat'] = data['hs_directory']['Location 1'].apply(lon_lat, False)
data['hs_directory']['lon'] = data['hs_directory']['Location 1'].apply(lon_lat)

# let us convert the new cols to numeric data
data['hs_directory']['lat'] = pd.to_numeric(data['hs_directory']['lat'], errors='coerce')
data['hs_directory']['lon'] = pd.to_numeric(data['hs_directory']['lon'], errors='coerce')
```

```python
# we will also the following columns in the ap_2010 to numeric_values
cols = ['AP Test Takers ', 'Total Exams Taken', 'Number of Exams with scores 3 4 or 5']

for col in cols: 
    data['ap_2010'][col] = pd.to_numeric(data['ap_2010'][col], errors='coerce')

data['ap_2010'].head(5)
```

# Condense datasets

## class_size

```python
# it is time to join the datasets to create a combined dataset. But we have to be careful as to which join we need
# to use. But before we do that, we need to condense some datasets. 


# We will start with class_size. 
# once again, here are the first few rows of class_size. 

data['class_size'].head(5)
```

**Observations: ** As seen, the first five rows all are associated with one school because the data is granularized by GRADE, PROGRAM TYPE and the next two columns. If we look at the unique values of GRADE, we find the following values. Since, we are only interested in the high school data, we are only going to select 9-12 GRADE data. From PROGRAM TYPE column, we will select the GED ED data as it is the largest category by far. 

```python
data['class_size'].loc[:, 'GRADE '].unique()
```

```python
# let us do the above wrangling
# for simplicity's sake, we will save the data['class_size'] into class_size

class_size = data['class_size']

# let us filter the class_size by GRADE first
class_size = class_size[class_size['GRADE '] == '09-12']

# let us now filter the PROGRAM_TYPE
class_size = class_size[class_size['PROGRAM TYPE'] == 'GEN ED']

# here are the first 3 rows of the class_size now.
class_size.head(3)
```

```python
# since we do not have a unique DBN in the above data, we are going to group the data by DBN.
class_size = class_size.groupby('DBN').agg(np.mean)
class_size = class_size.reset_index()

# let us save the class_size back to the original list of datasets
data['class_size'] = class_size
```

```python
data['class_size'].head(4)
```

## demographics

```python
# for the sake of simplicity
demographics = data['demographics']

demographics.head(5)
```

```python
# as we can see the only column that prevents the DBN from being unique is the schoolyear column. Let us condense
# the dataset so we are only focused on the schoolyear == 20112012
demographics = demographics[demographics['schoolyear'] == 20112012]

# let us save it back to the data['demographics']
data['demographics'] = demographics

# let us now take a look, as you can see we have different DBNs now.
data['demographics'].head(5)
```

## graduation

```python
data['graduation'].head(5)
```

```python
## we can see above that the Demographic and Cohort columns prevent the DBN from being unique. Let us choose the
# most recent cohort, which is 2006. And we will pick rows from the Total Cohort of the Demographic column.

# for the sake of simplicity
graduation = data['graduation']

# let us filter the Demographic col first. 
graduation = graduation[graduation['Demographic'] == 'Total Cohort']

# onto the Cohort col now. 
graduation = graduation[graduation['Cohort'] == '2006']

# saving it back to original
data['graduation'] = graduation

# first few rows now. 
data['graduation'].head(5)
```

# Combine the datasets

```python
# so let us start combining data sets to the combined. We will first combine sat_results with ap_2010
combined = data['sat_results']

# we will use the left merge strategy since other sets such as graduation and ap_2010 have many missing values. 
combined = pd.merge(combined, data['ap_2010'], on='DBN', how='left')

# let us merge the graduation data set to the combined also. 
combined = pd.merge(combined, data['graduation'], on='DBN', how='left')

# let us merge class_size into combined with inner merge
combined = pd.merge(combined, data['class_size'], on='DBN', how='inner')

# merging demographics now
combined = pd.merge(combined, data['demographics'], on='DBN', how='inner')

# merging survey now
combined = pd.merge(combined, data['survey'], on='DBN', how='inner')

# merging hs_directory now
combined = pd.merge(combined, data['hs_directory'], on='DBN', how='inner')

print (combined.shape)
combined.head(5)
```

```python
# let us derive the means of different columns of combined 
means = combined.mean()

# let us fill the na values with the means
combined = combined.fillna(means)

# let us fill any remaining na values with 0
combined = combined.fillna(0)
combined.head(5)
```

# Add a school district column for mapping

```python
# let us create the school_dist column for further analysis of our data
combined['school_dist'] = combined['DBN'].apply(lambda x: x[:2])
combined['school_dist'].head(5)
```
